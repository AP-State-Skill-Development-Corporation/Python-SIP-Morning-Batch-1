```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [5, 10, 2, 18, 6, 15, 8]

plt.plot(x, y, linestyle = '--', linewidth = 1.5, color = 'r', marker = '*', label = 'Lineplot')
plt.xlabel("X - axis")
plt.ylabel("Y - axis")
plt.title('Line Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

x = np.radians(np.arange(-360, 720, 15))
siny = np.sin(x)
cosy = np.cos(x)
plt.figure(figsize=(20, 10)) # w, h
plt.plot(x, siny, linestyle = '-.', linewidth = 1.5, color = 'b', marker = '*', label = 'Sin Plot')
plt.plot(x, cosy, linestyle = '--', linewidth = 1.5, color = 'r', marker = 'o', label = 'Cos Plot')
plt.legend()
plt.grid()
plt.show()
```

```python
help(plt.legend)
```

```python
import numpy as np

x = np.radians(np.arange(-360, 361, 15))
siny = np.sin(x)
cosy = np.cos(x)

plt.plot(x, siny, linestyle = '-.', linewidth = 1.5, color = 'b', marker = '*', label = 'Sin Plot')
plt.show()
plt.plot(x, cosy, linestyle = '--', linewidth = 1.5, color = 'r', marker = 'o', label = 'Cos Plot')
plt.legend()
plt.grid()
plt.show()
```

## Boxplot

```python
plt.figure(figsize=(15, 5))
box = [np.random.normal(0, i, 100) for i in range(0, 4)]
#print(box)
plt.boxplot(box)

plt.show()
```

```python
box.mean()
```

```python
box.max(), box.min()
```

```python
np.percentile(box, 75), np.percentile(box, 25)
```

## Pie Chart

```python
programming = ['Python', 'C', 'C++', 'Java', 'Reactjs']
pop = [30, 10, 15, 25, 20]
sum(pop)
```

```python
share = [0.4, 0.0, 0.0, 0.2, 0.0]
plt.pie(pop, labels = programming, explode=share, radius=1.5)
plt.show()
```
